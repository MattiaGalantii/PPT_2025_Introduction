Rabbit's population balance

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def rabbit_population(t, y, r, K):
    dydt = r * y * (1 - y / K)
    return dydt

# Parameters
r = 0.1  # growth rate
K = 100  # carrying capacity
y0 = [10]  # initial population
t_span = (0, 100)  # time span for the simulation
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Solve ODE
solution = solve_ivp(rabbit_population, 
                     t_span, 
                     y0, 
                     args=(r, K), 
                     t_eval=t_eval)

# Plot results
plt.plot(solution.t, solution.y[0], label='Rabbit Population', linewidth=3)
plt.plot(0, y0, 'ro', label='Initial Population', markersize=8)
plt.hlines(K, colors='g', linestyles='--', label='Carrying Capacity (K)', xmin=t_span[0], xmax=t_span[1], linewidth=2)
plt.title('Rabbit Population Growth Over Time')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.grid()
plt.show()
```
